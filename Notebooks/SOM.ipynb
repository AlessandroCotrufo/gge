```python
import gge
from gge import models
import numpy as np
```

```python
# Using the SOM class
test = np.load("/Users/kaaso/Documents/phd/coding/Landsat/data/ndwi_stack_test.npy")
test_flat = test.reshape(test.shape[0], -1)  # Automatically flatten the array
```

```python

som = models.SOM(num_neurons=(10, 10), input_dim=test_flat.shape[1], learning_rate = 0.0001, epochs=100)
som.train(test_flat)
```

```python
som.plot_u_matrix()
```

```python
activations = som.get_activations(test_flat)
anomalies = som.detect_anomalies(test_flat, threshold=1)
anomaly_coordinates = [som.index_to_xy(idx, test.shape[2]) for idx in anomalies]  # Assuming test.shape[2] is the width of your images
```

```python

models.plot_activations_on_image(test[0, :, :], activations, 10, anomaly_coordinates)
```
